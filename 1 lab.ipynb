Применение PCA для снижения размерности и избовления от мультиколлиниарности

```python
# Применение PCA
pca = PCA()
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Объясненная дисперсия
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

# График объясненной дисперсии
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.xlabel('Количество компонент')
plt.ylabel('Накопленная объясненная дисперсия')
plt.title('Накопленная объясненная дисперсия PCA')
plt.grid(True)
plt.show()

# Выбор количества компонент (например, 95% дисперсии)
n_components = np.argmax(cumulative_variance >= 0.95) + 1
print(f"Количество компонент для 95% дисперсии: {n_components}")

# Применение PCA с выбранным количеством компонент
pca = PCA(n_components=n_components)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

print(f"Размерность после PCA: {X_train_pca.shape}")

# Линейная регрессия на PCA
lr_pca = LinearRegression()
lr_pca.fit(X_train_pca, y_train)
y_pred_lr_pca = lr_pca.predict(X_test_pca)

# Гребневая регрессия на PCA
ridge_pca = Ridge(alpha=1.0)
ridge_pca.fit(X_train_pca, y_train)
y_pred_ridge_pca = ridge_pca.predict(X_test_pca)

# Оценка моделей на PCA
print("\nОценка моделей на PCA компонентах:")
lr_pca_metrics = evaluate_model(y_test, y_pred_lr_pca, "Линейная регрессия (PCA)")
ridge_pca_metrics = evaluate_model(y_test, y_pred_ridge_pca, "Гребневая регрессия (PCA)")

# Кросс-валидация на PCA
lr_pca_cv_scores = cross_val_score(lr_pca, X_train_pca, y_train, cv=5, scoring='r2')
ridge_pca_cv_scores = cross_val_score(ridge_pca, X_train_pca, y_train, cv=5, scoring='r2')

print(f"Кросс-валидация R² (линейная PCA): {lr_pca_cv_scores.mean():.4f}")
print(f"Кросс-валидация R² (гребневая PCA): {ridge_pca_cv_scores.mean():.4f})")
```

Благодаря PCA мы смогли снизить размерность до 10 признаков. Качества модели также - не плохие.
